```python
### Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.color_palette("Blues", as_cmap=True)

### Standard Packages
import numpy as np
import warnings
import pandas as pd
import os
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
warnings.filterwarnings("ignore")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('../data/subgenre_dataset_features.csv', index_col=0)
```

```python
df.head(103)
```

## Create 3 class dataset and train_test_split

```python
df3 = df.loc[(df['class'] == 'BigRoom') | (df['class'] == 'DrumAndBass') | (df['class'] == 'Techno')]
```

```python
X = df3.drop('class', axis=1)
y = df3['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=123, stratify=y)
```

```python
y_train.value_counts()
```

## Create Random Forest Classifier and Train

```python
rf = RandomForestClassifier(n_estimators=100, criterion='log_loss', max_depth=15, 
                            min_samples_split=2, n_jobs=8, verbose=1, random_state=123)
```

```python
rf.fit(X_train, y_train)
```

## RF Evaluation: Accuracy is 95% for BigRoom vs. DrumAndBass vs. Techno classification

```python
print('The accuracy is: ' + str(round(accuracy_score(rf.predict(X_test), y_test), 2)))
```
